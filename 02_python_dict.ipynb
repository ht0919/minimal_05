## ■(1) 宣言と初期化

```python
# 空の辞書を宣言
dic = {}
print('空の辞書 dic =', dic)

# 宣言時に初期値を設定
dic = {'apple':100, 'orange':200, 'melon':300} # {'キー1':値1, 'キー2':値2,…}
print('初期値を設定 dic =', dic)
```

## ■(2) 参照

```python
# すべてのデータを参照
print('すべてのデータ dic =', dic)

# キーを指定して参照
print('キーが orange の値 =', dic['orange']) # dic['キー']で参照

# キーが見つからない場合はエラーになります。
```

## ■(3) 値の更新(上書き)

```python
# すべてのデータを参照
print('すべてのデータ dic =', dic)

# 既存のキーの値を更新
dic['melon'] = 500
print('melonの値を500に更新 dic =', dic)

# 更新した値を元に戻す
dic = {'apple':100, 'orange':200, 'melon':300} # {'キー1':値1, 'キー2':値2,…}
```

## ■(4) 要素の削除

```python
# すべてのデータを参照
print('すべてのデータ dic =', dic)

# 指定の要素を削除
del dic['apple']
print('キーがapple要素を削除 dic =', dic)

# 削除した値を元に戻す
dic = {'apple':100, 'orange':200, 'melon':300} # {'キー1':値1, 'キー2':値2,…}
```

## ■(5) 要素の追加

```python
# すべてのデータを参照
print('すべてのデータ dic =', dic)

# 新しい要素を追加
dic['banana'] = 250
print('キーがbananaの要素を追加 dic =', dic)

# 複数の要素をまとめて追加(別の辞書と結合)
dic.update({'lemon':50, 'kiwi':250})
print('複数の要素を追加 dic =', dic)

# 追加した要素を元に戻す
dic = {'apple':100, 'orange':200, 'melon':300} # {'キー1':値1, 'キー2':値2,…}
```

## ■(6) 集計情報の取得

```python
# すべてのデータを参照
print('すべてのデータ dic =', dic)

# 辞書の要素数
print('要素数:', len(dic))
# 辞書の合計値
print('合計値:', sum(dic.values()))
# 辞書の平均値
print('平均値:', sum(dic.values())/len(dic))
# 辞書の最大値
print('最大値:', max(dic.values()))
# 辞書の最小値
print('最小値:', min(dic.values()))
```

## ■(7) 繰り返しによる辞書の操作

```python
# すべてのデータを参照
print('すべてのデータ dic =', dic)

# for文ですべての要素を取り出す
for key, value in dic.items():
  print(key, value)
```
